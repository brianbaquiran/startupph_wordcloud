```python
with open("startup_ph_corpus.txt") as f:
    raw = f.read()
```

```python
from nltk.tokenize import casual_tokenize, word_tokenize
```

```python
casual_tokenize(raw)
```

```python
from string import punctuation
```

```python
from nltk.corpus import stopwords

sw = set(stopwords.words('english') + list(punctuation))
```

```python
sw
```

```python
words = casual_tokenize(raw)
```

```python
words = [w.lower() for w in words if w not in sw and not w.isdigit() and len(w) > 3 and '@' not in w]
```

```python
words
```

```python
word_tokenize(" ".join(words))
```
